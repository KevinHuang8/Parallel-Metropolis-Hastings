```python
import matplotlib.pyplot as plt
```

```python
x = []
y = []
with open('samples_gpu.txt', 'r') as f:
    for line in f:
        sample = [float(i) for i in line.split()]
        x.append(sample[0])
        y.append(sample[1])
```

```python
plt.scatter(x, y, s = 3, alpha=0.0075)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.savefig('output_plot.png')
```
